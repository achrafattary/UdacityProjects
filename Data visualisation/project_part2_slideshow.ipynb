## Investigation Overview

the goal of this project is to study whether certain variable like age and user type affect how the users uses this service

## Dataset Overview

this data set contains rides from february of 2019,the dataset have many informations both about the rides(duration,time_start,end_time,location_start ect) and about the users (age,user_type,ect)

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
df = pd.read_csv("dataset_final.csv")
```

## Visualization 1

>this visual show that the vast majority of the user base is subscribed,the retention rate must be very hight and the user satisfied

```python
def ratio_user_pie() : 
    count = df["user_type"].value_counts()
    labels=["Subscriber","Customer"]
    plt.figure(figsize= (8,8))
    plt.title("Ration of customers and subscribers")
    plt.pie(count,labels=labels,startangle=90,autopct='%1.3f %%');
    plt.show()
```

```python
ratio_user_pie()
```

## Visualization 2

here we can see that whether the user is subscribed or not,they still use the service at the same hour every day with a peak at 8 in the morning and 18 probably because workers use the service to get home



```python
color = sb.color_palette()[0]
def number_customers() :
    hours = [v for v in range (0,24)]
    g = sb.catplot(data=df, x='start_hour', col="user_type", kind='count', sharey = False,order=hours,color=color)
    g.set_xlabels('hour in 24h format')
    g.set_ylabels('number of rider')
    g.fig.subplots_adjust(top=0.7)
    g.fig.suptitle('numbers of rides depending on the hour and user_type',size = 20)
```

```python
number_customers()
```

## Visualization 3

in this visual we can see that users folow the same pattern form the duration of their ride,especially the users in the age of work,this also make us think that most users use the service to move from their jobs to their work and vise-versa

```python
def distribution_age_duration():  
    plt.figure(figsize= (7,7))
    sb.violinplot(data=df, y='membre_age_range', x='start_hour', inner='quartile',color = color)
    plt.xlabel('Duration in seconds')
    plt.ylabel('age range')
    plt.title('Duration depending on age range');
```

```python
distribution_age_duration()
```

## Visualization 4

this visual also shows that gender does not make a difference in the pattern of usage of the service

```python
def FaceGrid_gender_age_duration():
    g = sb.FacetGrid(data = df, col = 'member_gender',height = 3.5,col_wrap=2)
    g.map(plt.scatter, 'membre_age', 'duration_sec',alpha = 0.5)
    g.set_xlabels('Member Age',size = 20)
    g.set_ylabels('Duration in seconds',size = 20)
    g.fig.subplots_adjust(top=0.9)
    g.fig.suptitle('duration of ride depending on members age and gender',size = 20)
```

```python
FaceGrid_gender_age_duration()
```
